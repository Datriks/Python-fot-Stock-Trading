## Time Series Modelling

```python
# decomposing time series
# import libraries
import pandas as pd 
import quandl
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# extract the price of gold from Quandl
df = quandl.get('WGC/GOLD_MONAVG_USD',
                start_date = '2011-1-1',
                end_date = '2020-6-10',
                api_key = 'UkBXz6sLeET-4uRJ-65q')
df.head()
```

```python
# rename the columns 
df.rename(columns={'Value':'price'},inplace=True)
# will get the last price of gold for each month
df = df.resample('M').last()
```

```python
print(f'Shape of Data Frame : {df.shape}')
df.head()
```

```python
# adding rolling mean and standard deviation()
df['rolling_mean'] = df.price.rolling(window=12).mean()
# gold volatility
df['rolling_std'] = df.price.rolling(window=12).std()
df.tail()
```

```python
#plot the findings
df.plot(title='Gold Price')
plt.show()
```

```python
# seasonal decomposition using multiplicative model:
decomp_results = seasonal_decompose(df.price,model='multiplicative')
# plot the results
decomp_results.plot().suptitle('Multiplicative Decomposition', 
                               fontsize=14)
plt.show()
```
